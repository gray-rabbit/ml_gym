```python
# import tensorflow as tf
import gym
import tensorflow as tf
import matplotlib.pyplot as plt
from keras import Sequential, layers, losses
import numpy as np
```

```python
env = gym.make("FrozenLake-v1", is_slippery=True)
```

```python
def get_model(input_shape):
    ni = tf.keras.layers.Input(shape=(input_shape))
    nn = tf.keras.layers.Dense(4, activation=None,kernel_regularizer=tf.random_uniform_initializer(0,0.01))(ni)
    model = tf.keras.Model(ni,nn)
    return model
def one_hot(s):
    m  = np.zeros((1,16))
    m[0,s] = 1
    return m
m = get_model((None, 16))
```

```python
dis,rList=.9, []
model = get_model((None,16))
opti = tf.keras.optimizers.SGD(learning_rate=0.1)
for i in range(20000):
    state = env.reset()
    game_over = False
    e = 1./((i/50)+10)
    while not game_over:
        predict = model(one_hot(state)).numpy()
        if np.random.rand(1)<e:
            action = env.action_space.sample()
        else:
            action = np.argmax(predict,1)[0]
        new_state,reward, game_over,_ = env.step(action)
        new_predict = model(one_hot(new_state)).numpy()
        maxValue = np.max(new_predict)
        if game_over:
            predict[0,action] = reward
        else:
            predict[0,action] =reward + dis*maxValue
        with tf.GradientTape() as tape:
            _pred = model(one_hot(state))
            _loss = tf.keras.losses.MeanSquaredError()(_pred, predict)
            # print(_pred, predict)
        grad = tape.gradient(_loss,model.trainable_variables)
        opti.apply_gradients(zip(grad, model.trainable_variables))
        state = new_state
    if reward:print(f"{i} reward:{reward}")
    rList.append(reward)
np.sum(rList)/20000
```

```python
counter = 0 
for i in range(100):
    game_over = False
    state = env.reset()
    while not game_over:
        predict = model.predict(one_hot(state))
        # env.render()
        action = np.argmax(predict,1)[0]
        new_state, reward, game_over, _ = env.step(action)
        state = new_state
        if game_over:
            # env.render()
            counter+=reward
    print(f"{i} : counter {counter}")
counter
```
